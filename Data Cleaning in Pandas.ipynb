```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\user\Downloads\Customer Call List.xlsx")
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

```python
# # Replace various representations of missing values with a blank space
# df['Phone_Number'] = df['Phone_Number'].replace(["N/a", "NaN", "nan", None, pd.NA], "", regex=False)

# # Replace any delimiters ('/', '|') in phone numbers with a standard '-' delimiter
# df['Phone_Number'] = df['Phone_Number'].str.replace(r"[|/]", "-", regex=True)

# # Ensure only valid numeric characters and '-' are retained in the Phone_Number column
# df['Phone_Number'] = df['Phone_Number'].str.replace(r"[^\d\-]", "", regex=True)

# # Display the cleaned DataFrame
# df


# Replace various representations of missing values with a blank space
df['Phone_Number'] = df['Phone_Number'].replace(["N/a", "NaN", "nan", None, pd.NA], "", regex=False)

# Ensure only numeric values are retained and remove invalid characters
df['Phone_Number'] = df['Phone_Number'].str.replace(r"[^\d]", "", regex=True)

# Add hyphens to phone numbers with valid 10 digits
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: f"{x[:3]}-{x[3:6]}-{x[6:]}" if len(x) == 10 else x)

# Display the cleaned DataFrame
df
```

```python

# # Adjusting the regex to capture missing state and zip code more gracefully
# df[['Street_address', 'State', 'Zip_code']] = df['Address'].str.extract(
#     r'^(.*?)(?:,\s*([^,]+))?(?:,\s*(\d+))?$'
# )
# df
df[['Street_address', 'State', 'Zip_code']] = df['Address'].str.extract(
    r'^(.*?)(?:,\s*([^,]+))?(?:,\s*(\d{5}))?$'
)
df
```

```python
# df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')

# df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
# df

# Replace "Yes" with "Y" in the "Do_Not_Contact" column
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes', 'Y', regex=False)

# Replace "No" with "N" in the "Do_Not_Contact" column
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No', 'N', regex=False)

# Display the updated DataFrame
df
```

```python
df = df.replace('N/a', '')
df = df.replace('n/a', '')  # Optional: Handle lowercase variations
df = df.fillna('')
df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)

df
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)

df
```

```python
df = df.reset_index(drop=True)
df
```
